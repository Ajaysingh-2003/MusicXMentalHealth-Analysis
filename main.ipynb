## Effects of Music on Mental Health Analysis

### Music Therapy

##### Music Therapy is the process to positively effect the mental health of an indiviual using music. It can result in decrease of anxiety, sadness, depression and many other mental health problems.

##### There are many factors and cases that show diffrent effect of music therapy on different indiviuals. Hence, here we are to study the interdependece or correlation between the various factors and the mental problems of a person.

#### Interpreting data
##### Block 0: Background
##### Respondents answer generic questions focused on musical background and listening habits.

##### Block 1: Music genres
##### Respondents rank how often they listen to 16 music genres, where they can select:
<ul>
<li>Never</li>
<li>Rarely</li>
<li>Sometimes</li>
<li>Very frequently</li>
</ul>

##### Block 2: Mental health
##### Respondents rank Anxiety, Depression, Insomnia, and OCD on a scale of 0 to 10, where:

##### 0 - I do not experience this.
##### 10 - I experience this regularly, constantly/or to an extreme.

## Building the model

```python
#import the machine learning libraries from scikit learn
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix as cm
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier , DecisionTreeRegressor, export_graphviz
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier , AdaBoostClassifier ,RandomForestRegressor , AdaBoostRegressor
from xgboost import XGBClassifier , XGBRegressor
from sklearn.linear_model import LinearRegression ,LogisticRegression, Ridge, RidgeCV, RidgeClassifier, RidgeClassifierCV, Lasso
from sklearn.neighbors import KNeighborsRegressor , KNeighborsClassifier
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV
from sklearn.metrics import mean_squared_error , mean_absolute_error , r2_score , classification_report , accuracy_score , f1_score , precision_score
import pickle
```

### Splitting the Data

```python
X=df.drop('Music effects',axis=1)
y=df[['Music effects']]
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.4,random_state=1)
```

```python
X.head()
```

```python
y.head()
```

```python
print(X_train.shape)
print(y_train.shape)
```

### Split the Categorical and numerical data

```python
object_cols = df.select_dtypes(include="object").columns.tolist()
num_cols = df.select_dtypes(include="number").columns.tolist()
```

```python

print("Numeric Columns:",num_cols)
print("Categorical Columns:",object_cols)
```

### Feature Scaling

```python
Standard_Scaler=StandardScaler()
X_train[num_cols] = Standard_Scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = Standard_Scaler.transform(X_test[num_cols])
```

### Encoding the values

```python
label_encoder = LabelEncoder() #To convert Categorical Data to Numeric Data
y_train = label_encoder.fit_transform(y_train)
object_columns = X_train.select_dtypes(include=['object']).columns
for column in object_columns:
    X_train[column] = label_encoder.fit_transform(X_train[column])
    
y_test = label_encoder.fit_transform(y_test)
object_columns = X_test.select_dtypes(include=['object']).columns
for column in object_columns:
    X_test[column] = label_encoder.fit_transform(X_test[column])
```

```python
X_train.head()
```

### Model Building using Scikit-Learn

```python
#Linear Regression Model
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_lrpred=regressor.predict(X_test)
acc=r2_score(y_test,y_lrpred)
coefficeient=regressor.coef_
print("The accuracy of the linear regression model is:",np.round(acc*100,2))
print("The coefficients are:",coefficeient)
```

#### We can observe that we are handling a categorical oriented datadet where we have to predict the outcome based on the categories of effect on mental health. So, in such cases Regression models will provide very less accuracy and won't provide satisfactory results.

#### Hence, we will be using Classifier models from now onwards.

### Ridge Classification

```python
#Ridge Classification Model
alphas=[0.01,0.1,1,10,1000]
for alpha in alphas:
    ridgecl=RidgeClassifier(alpha=alpha,max_iter=100,solver='auto')
    ridgecl.fit(X_train,y_train)
    y_ridgepred=ridgecl.predict(X_test)
    ridgeacc=accuracy_score(y_test,y_ridgepred)
print("The accuracy of Ridge Classification is:",np.round(ridgeacc*100,2))
print(classification_report(y_test,y_ridgepred))
```

#### Ridge Classification provided us a great accuracy of 73.6%. But we can try various other models that may provide greater accuracy scores.

### Logistic Regression

#### Logistic Regression is a type of classification model which is used for binary classification of the outcome.

```python
logis=LogisticRegression()
logis.fit(X_train,y_train)
y_logis_pred=logis.predict(X_test)
logis_acc=accuracy_score(y_test,y_logis_pred)
print("The accuracy for logistic regression is:",np.round(logis_acc*100,2))
print(classification_report(y_test,y_logis_pred))
```

### Decision Tree Classifier

```python
dtc = DecisionTreeClassifier(max_depth = 10, random_state=1)
dtc.fit(X_train, y_train)
y_dtc_pred = dtc.predict(X_test)
dtc_acc=accuracy_score(y_test,y_dtc_pred)
print("The accuracy of Decision tree Classifier is:",np.round(dtc_acc*100,2))
print(classification_report(y_test,y_dtc_pred))
```

### Visualising the Decision Tree Algorithm

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtc,
                   filled=True)
```

### Random Forest Classifier

```python
rfc=RandomForestClassifier(n_estimators=100,random_state=30)
rfc.fit(X_train,y_train)
y_rfc_pred=rfc.predict(X_test)
rfc_acc=accuracy_score(y_test,y_rfc_pred)
print("The accuracy of Random Forest Classifier is:",np.round(rfc_acc*100,2))
```

### KNeighbour Classifier

```python
kneigh=KNeighborsClassifier(n_neighbors=1)
kneigh.fit(X_train,y_train)
y_kn_pred=kneigh.predict(X_test)
kn_acc=accuracy_score(y_test,y_kn_pred)
print("The accuracy score of Kneighbours classifier is:",np.round(kn_acc*100,2))
```

### Analysing the accuracy of the models

```python
models={"Ridge Classification":ridgeacc*100,
        "LogisticRegression":logis_acc*100,
        "Decision Tree Classification":dtc_acc*100,
        "RandomForestClassifier":rfc_acc*100,
        "Kneighbours Classification":kn_acc*100}
models = {k:[v] for k,v in models.items()}
df_accuracy=pd.DataFrame(models)
```

```python
sns.set_style('whitegrid')
fig=plt.figure(figsize=(7,4))
sns.barplot(df_accuracy)
plt.tight_layout()
plt.xlabel("ML Models")
plt.ylabel("Accuracy of Models")
plt.title("Accuracy of various ML Models")
plt.xticks(rotation=35)
plt.show()
```

### From the above barplot, we can conclude that the Ridge Classification model provides us the best accuracy on the dataset.
